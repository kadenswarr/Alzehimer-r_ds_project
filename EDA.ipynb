```python
category_names  = ["No High School Diploma", "High School Grad", "Some Collge", "Bachelor's degree"]

results  = {
    "Less than high school graduate" : df["Percent!!Population 18 to 24 years!!Less than high school graduate"],
    "High school graduate" : df["Percent!!Population 18 to 24 years!!High school graduate (includes equivalency)"],
    "Some college or associate's degree" : df["Percent!!Population 18 to 24 years!!Some college or associate's degree"],
    "Bachelor's degree or higher" : df["Percent!!Population 18 to 24 years!!Bachelor's degree or higher"],
}

rest_results = {
    "Less than High School" : df['Percent!!Population 25 years and over!!Less than 9th grade'],
    "Some High School" : df["Percent!!Population 25 years and over!!9th to 12th grade, no diploma"],
    "High School Graduate" : df['Percent!!Population 25 years and over!!High school graduate (includes equivalency)'],
    "Some College" : df['Percent!!Population 25 years and over!!Some college, no degree'],
    "Associate's Degree" : df['Percent!!Population 25 years and over!!Associate\'s degree'],
    "Bachelor's Degree" : df['Percent!!Population 25 years and over!!Bachelor\'s degree'],
    "Gradute or Proffessional Degree" : df['Percent!!Population 25 years and over!!Graduate or professional degree']
}


def education(results, category_names, title):
    labels = list(df.index)
    data = np.array(list(results.values()))
    data_cum = np.cumsum(data, axis=0)
    
    
    fig, ax = plt.subplots(figsize=(20, 15))
    ax.set_title(title, loc="center")
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, 100)
    cat_colors = plt.get_cmap('BrBG')(
                np.linspace(0.15, 0.85 , data.shape[0]))

    for i, (colname, color) in enumerate(zip(category_names, cat_colors)):
        state = []
        state_cum = []
        for j in zip(range(data.shape[1])):
            state.append(data[i,j][0])
            state_cum.append(data_cum[i,j][0])
        widths = np.array(state)
        starts = np.array(state_cum) - widths
        xcenters = starts + widths / 2 

        ax.barh(labels, widths, left=starts, height=0.5,
               label=colname, color=color)


        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(round(c, 2)), ha="center", va="center", color="black")

    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 0),
            loc="lower left")

    return fig, ax

education(results, category_names, "18 to 24 Years Old Education Percent")
plt.show()
education(rest_results, list(rest_results.keys()), "25 Years Old and Over Education Percent")
plt.show()
```
